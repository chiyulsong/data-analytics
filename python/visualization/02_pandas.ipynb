# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [23, 45, 56, 78],
    'Value2': [19, 29, 39, 49],
    'Value3': [35, 25, 45, 65]
}

df = pd.DataFrame(data)

# Create subplot (4 rows x 3 columns)
fig, axes = plt.subplots(4, 3, figsize=(15, 20))

# Chart 1: Bar Chart
df.plot(kind='bar', x='Category', y='Value1', ax=axes[0, 0], title='Bar Chart')

# Chart 2: Line Chart
df.plot(kind='line', x='Category', y='Value2', ax=axes[0, 1], title='Line Chart')

# Chart 3: Scatter Plot
df.plot(kind='scatter', x='Value1', y='Value2', ax=axes[0, 2], title='Scatter Plot')

# Chart 4: Horizontal Bar Chart
df.plot(kind='barh', x='Category', y='Value3', ax=axes[1, 0], title='Horizontal Bar Chart')

# Chart 5: Box Plot
df[['Value1', 'Value2', 'Value3']].plot(kind='box', ax=axes[1, 1], title='Box Plot')

# Chart 6: Area Chart
df.plot(kind='area', x='Category', y='Value1', ax=axes[1, 2], title='Area Chart')

# Chart 7: Pie Chart
df.set_index('Category')['Value1'].plot(kind='pie', ax=axes[2, 0], title='Pie Chart')

# Chart 8: Histogram
df['Value1'].plot(kind='hist', ax=axes[2, 1], title='Histogram')

# Chart 9: Density Plot
df['Value2'].plot(kind='density', ax=axes[2, 2], title='Density Plot')

# Chart 10: Hexbin Plot (Note: requires numerical x and y)
df.plot(kind='hexbin', x='Value1', y='Value2', gridsize=25, ax=axes[3, 0], title='Hexbin Plot')

# Chart 11: KDE Plot
df['Value3'].plot(kind='kde', ax=axes[3, 1], title='KDE Plot')

# Chart 12: Custom Plot (e.g., Step Plot)
axes[3, 2].step(df['Category'], df['Value1'], where='mid')
axes[3, 2].set_title('Step Plot')

# Adjust layout
plt.tight_layout()
plt.show()
```
