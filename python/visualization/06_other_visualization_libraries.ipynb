```python
import altair as alt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
df = pd.DataFrame({
    'x': np.random.randn(500),
    'y': np.random.randn(500),
    'category': np.random.choice(['A', 'B', 'C'], 500)
})

# Define the brush
brush = alt.selection_interval()

# Scatter plot with brush
scatter = alt.Chart(df).mark_point().encode(
    x='x:Q',
    y='y:Q',
    color=alt.condition(brush, 'category:N', alt.value('lightgray')),
    tooltip=['x', 'y', 'category']
).add_selection(
    brush
).properties(
    width=400,
    height=400,
    title='Scatter Plot with Brushing'
)

# Histogram for x with brush
hist_x = alt.Chart(df).mark_bar().encode(
    x=alt.X('x:Q', bin=True),
    y='count():Q',
    color=alt.condition(brush, 'category:N', alt.value('lightgray'))
).transform_filter(
    brush
).properties(
    width=400,
    height=400,
    title='Histogram of X with Brushing'
)

# Histogram for y with brush
hist_y = alt.Chart(df).mark_bar().encode(
    x=alt.X('y:Q', bin=True),
    y='count():Q',
    color=alt.condition(brush, 'category:N', alt.value('lightgray'))
).transform_filter(
    brush
).properties(
    width=400,
    height=400,
    title='Histogram of Y with Brushing'
)

# Combine charts into subplots
interactive_chart = alt.vconcat(
    scatter,
    hist_x,
    hist_y
).resolve_scale(
    color='independent'
)

interactive_chart.show()
```
