# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Figure  
fig, axs = plt.subplots(4, 3, figsize=(20, 20))  # 4 3 

# 1.   (Line Plot)
data_line = pd.DataFrame({
    'date': pd.date_range(start='1/1/2021', periods=100),
    'value': np.random.randn(100).cumsum()
})
data_line.plot(x='date', y='value', title='Line Plot', ax=axs[0, 0])
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Value')

# 2.   (Bar Plot)
data_bar = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D', 'E'],
    'value': [3, 7, 5, 6, 4]
})
data_bar.plot(kind='bar', x='category', y='value', title='Bar Plot', ax=axs[0, 1])
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')

# 3.  (Histogram)
data_hist = pd.DataFrame({
    'value': np.random.randn(1000)
})
data_hist.plot(kind='hist', y='value', bins=30, title='Histogram', ax=axs[0, 2])
axs[0, 2].set_xlabel('Value')
axs[0, 2].set_ylabel('Frequency')

# 4.  (Scatter Plot)
data_scatter = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50)
})
data_scatter.plot(kind='scatter', x='x', y='y', title='Scatter Plot', ax=axs[1, 0])
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# 5.   (Box Plot)
data_box = pd.DataFrame({
    'category': ['A']*50 + ['B']*50 + ['C']*50,
    'value': np.random.randn(150)
})
data_box.boxplot(by='category', column=['value'], grid=False, ax=axs[1, 1])
axs[1, 1].set_title('Box Plot')
axs[1, 1].set_xlabel('Category')
axs[1, 1].set_ylabel('Value')
fig.suptitle('')  #     

# 6.   (Pie Chart)
data_pie = pd.Series([30, 20, 25, 25], index=['A', 'B', 'C', 'D'], name='Value')
data_pie.plot(kind='pie', title='Pie Chart', autopct='%1.1f%%', ax=axs[1, 2])
axs[1, 2].set_ylabel('')  # Y  

# 7.   (Density Plot)
data_density = pd.DataFrame({
    'value': np.random.randn(1000)
})
data_density.plot(kind='density', title='Density Plot', ax=axs[2, 0])
axs[2, 0].set_xlabel('Value')
axs[2, 0].set_ylabel('Density')

# 8.   (Area Plot)
data_area = pd.DataFrame({
    'date': pd.date_range(start='1/1/2021', periods=100),
    'value1': np.random.rand(100).cumsum(),
    'value2': np.random.rand(100).cumsum()
})
data_area.set_index('date').plot(kind='area', title='Area Plot', ax=axs[2, 1])
axs[2, 1].set_xlabel('Date')
axs[2, 1].set_ylabel('Value')

# 9.    (Stacked Bar Plot)
data_stacked_bar = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D', 'E'],
    'value1': [3, 7, 5, 6, 4],
    'value2': [2, 4, 6, 8, 5]
})
data_stacked_bar.set_index('category').plot(kind='bar', stacked=True, title='Stacked Bar Plot', ax=axs[2, 2])
axs[2, 2].set_xlabel('Category')
axs[2, 2].set_ylabel('Value')

# 10.  (Heatmap)
data_heatmap = pd.DataFrame(np.random.rand(10, 12))
cax = axs[3, 0].imshow(data_heatmap, cmap='viridis', aspect='auto')
fig.colorbar(cax, ax=axs[3, 0])
axs[3, 0].set_title('Heatmap')

#       
fig.delaxes(axs[3, 1])
fig.delaxes(axs[3, 2])

#   
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [23, 45, 56, 78],
    'Value2': [19, 29, 39, 49],
    'Value3': [35, 25, 45, 65]
}

df = pd.DataFrame(data)

# Create subplot (4 rows x 3 columns)
fig, axes = plt.subplots(4, 3, figsize=(15, 20))

# Chart 1: Bar Chart
df.plot(kind='bar', x='Category', y='Value1', ax=axes[0, 0], title='Bar Chart')

# Chart 2: Line Chart
df.plot(kind='line', x='Category', y='Value2', ax=axes[0, 1], title='Line Chart')

# Chart 3: Scatter Plot
df.plot(kind='scatter', x='Value1', y='Value2', ax=axes[0, 2], title='Scatter Plot')

# Chart 4: Horizontal Bar Chart
df.plot(kind='barh', x='Category', y='Value3', ax=axes[1, 0], title='Horizontal Bar Chart')

# Chart 5: Box Plot
df[['Value1', 'Value2', 'Value3']].plot(kind='box', ax=axes[1, 1], title='Box Plot')

# Chart 6: Area Chart
df.plot(kind='area', x='Category', y='Value1', ax=axes[1, 2], title='Area Chart')

# Chart 7: Pie Chart
df.set_index('Category')['Value1'].plot(kind='pie', ax=axes[2, 0], title='Pie Chart')

# Chart 8: Histogram
df['Value1'].plot(kind='hist', ax=axes[2, 1], title='Histogram')

# Chart 9: Density Plot
df['Value2'].plot(kind='density', ax=axes[2, 2], title='Density Plot')

# Chart 10: Hexbin Plot (Note: requires numerical x and y)
df.plot(kind='hexbin', x='Value1', y='Value2', gridsize=25, ax=axes[3, 0], title='Hexbin Plot')

# Chart 11: KDE Plot
df['Value3'].plot(kind='kde', ax=axes[3, 1], title='KDE Plot')

# Chart 12: Custom Plot (e.g., Step Plot)
axes[3, 2].step(df['Category'], df['Value1'], where='mid')
axes[3, 2].set_title('Step Plot')

# Adjust layout
plt.tight_layout()
plt.show()
```
